```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
time = 2*np.pi*(2*np.arange(1000)/1000 - 1)
signal = np.sin(10*time)
```

```python
plt.plot(time, signal)
plt.show()
```

```python
spectrum = np.fft.fft(signal)
```

```python
plt.plot(abs(spectrum))
plt.show()
```

```python
spectrum = np.zeros(1000, dtype=np.complex)
spectrum[2] = -1j
spectrum[1000-2] = 1j
```

```python
plt.plot(np.imag(spectrum))
plt.show()
```

```python
signal = np.fft.ifft(spectrum)
```

```python
plt.plot(time, np.real(signal), time, np.imag(signal))
plt.show()
```

```python
signal = np.exp(-(2*time)**2)
plt.plot(time, signal)
plt.show()
```

```python
noisy_signal = signal + 0.05*np.random.normal(size=1000)
plt.plot(time, noisy_signal)
plt.show()
```

```python
spectrum = np.fft.fft(noisy_signal)
plt.plot(time, abs(spectrum))
plt.show()
```

```python
filtered_spectrum = np.copy(spectrum)
filtered_spectrum[abs(spectrum) < 5] = 0
plt.plot(time, abs(filtered_spectrum))
plt.show()
```

```python
filtered_signal = np.fft.ifft(filtered_spectrum)
plt.plot(time, np.real(filtered_signal))
plt.show()
```

```python
diff = np.sum(np.abs(signal - np.real(filtered_signal)))/(time[-1] - time[0])
diff
```
